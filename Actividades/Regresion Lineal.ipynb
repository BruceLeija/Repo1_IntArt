```python
#Actividad 3: Solución de problemas. N4 006
#Integrantes del equipo
#Bayron Javier Hernández Sifuentes 2132169
#Brisa Marlene Mar Santiago 2127823
#Leonardo Rodrigo Salinas Mejia 2035376
#Bruce Brandon Leija Cortés 1982141
#librerias
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Datos La relación entre las horas de estudio y las calificaciones se establece con una fórmula lineal, y se agrega un término de error aleatorio
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 3 * horas_estudio + np.random.randn(100, 1) * 2

X_train, X_test, y_train, y_test = train_test_split(horas_estudio, calificaciones, test_size=0.32, random_state=42)

# Entrenamiento del modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Prueba del modelo
y_pred = model.predict(X_test)

# Calcular el error Se calcula el error cuadrático medio comparando las predicciones del modelo con las verdaderas calificaciones utilizando la función mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Error del cuadrático medio:", mse)

# Graficar el modelo de regresión lineal con las predicciones
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación')
plt.title('Modelo de Regresión Lineal')
plt.show()
```
